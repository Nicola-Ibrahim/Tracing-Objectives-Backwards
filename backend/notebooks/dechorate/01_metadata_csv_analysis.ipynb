# Analyze dEchorate Metadata (CSV)
Load the CSV metadata, filter by room/source/signal, and prepare outputs for the HDF5 notebook.


```python
# 0) Imports
from pathlib import Path
import numpy as np
import pandas as pd
```

## 1) Configuration


```python
# 1) User config
METADATA_PATH = Path("../../data/dEchorate/raw/dEchorate_database.csv")
PROCESSED_METADATA_PATH = Path(
    "../../data/dEchorate/processed/dEchorate_database_cleaned.csv"
)
```

```python
# 2) Helpers: metadata loading + light validation
def load_metadata(path: Path) -> pd.DataFrame:
    if not path.exists():
        raise FileNotFoundError(f"Metadata file not found: {path.resolve()}")

    if path.suffix.lower() != ".csv":
        raise ValueError(
            f"Unsupported metadata file type: {path.suffix}. Expected .csv"
        )

    df = pd.read_csv(path, sep=None, engine="python")
    df.columns = [str(c).strip() for c in df.columns]

    drop_cols = [c for c in df.columns if c == "" or str(c).startswith("Unnamed")]
    if drop_cols:
        df = df.drop(columns=drop_cols)

    print(f"Loaded metadata: {path.name} | rows={len(df):,} cols={len(df.columns)}")
    return df


def validate_numeric_columns(df: pd.DataFrame, cols: list[str]) -> dict:
    """Check for missing values, outliers, and invalid entries in numeric columns"""
    report = {}
    for col in cols:
        if col not in df.columns:
            report[col] = {"status": "missing"}
            continue

        col_data = pd.to_numeric(df[col], errors="coerce")
        report[col] = {
            "missing_count": col_data.isna().sum(),
            "missing_pct": col_data.isna().mean() * 100,
            "min": col_data.min(),
            "max": col_data.max(),
            "mean": col_data.mean(),
            "std": col_data.std(),
        }
    return report


def validate_positional_data(df: pd.DataFrame) -> pd.DataFrame:
    """Ensure source/mic positions are within room boundaries"""
    position_cols = [
        "src_pos_x",
        "src_pos_y",
        "src_pos_z",
        "mic_pos_x",
        "mic_pos_y",
        "mic_pos_z",
    ]

    # Convert string-like numbers to float
    for col in position_cols:
        df[col] = (
            pd.to_numeric(
                df[col].astype(str).str.replace(".", "", regex=False), errors="coerce"
            )
            / 1e8
        )  # Fix formatting artifacts if any
        # Note: If your data is already clean floats, use simple pd.to_numeric(df[col], errors='coerce')

    df["position_valid"] = True
    # Basic check for non-negative coordinates
    for axis in ["x", "y", "z"]:
        src_col = f"src_pos_{axis}"
        mic_col = f"mic_pos_{axis}"
        df["position_valid"] &= (df[src_col] >= 0) & (df[mic_col] >= 0)

    return df


def compute_src_mic_distance(df: pd.DataFrame) -> pd.DataFrame:
    """Calculate Euclidean distance between source and microphone"""
    df["src_mic_distance"] = np.sqrt(
        (df["src_pos_x"] - df["mic_pos_x"]) ** 2
        + (df["src_pos_y"] - df["mic_pos_y"]) ** 2
        + (df["src_pos_z"] - df["mic_pos_z"]) ** 2
    )
    return df
```

## 2) Load and Inspect


```python
# Load metadata
metadata_df = load_metadata(METADATA_PATH)
print(
    "Metadata columns (first 30): "
    + ", ".join(map(str, list(metadata_df.columns)[:30]))
)
print("Preview:")
display(metadata_df.head(5))
```

## 3) Filter Metadata


```python
# 3) Filter metadata
ROOM_CODES = [0, 1, 2]
SRC_IDS = [0, 1, 2]
SIGNAL_NAMES = ["chirp"]

required_cols = [
    "room_code",
    "src_id",
    "src_signal",
    "mic_id",
    "mic_pos_x",
    "mic_pos_y",
    "mic_pos_z",
]
missing = [c for c in required_cols if c not in metadata_df.columns]
if missing:
    raise KeyError(f"Missing required columns: {missing}")

subset_df = metadata_df.query(
    "room_code in @ROOM_CODES and src_id in @SRC_IDS and "
    "src_signal.str.lower() in @SIGNAL_NAMES"
).copy()

print(
    f"Filtered case -> rooms={ROOM_CODES}, src={SRC_IDS}, sig={SIGNAL_NAMES} "
    f"| rows={len(subset_df)}"
)


needed_cols = [
    "room_code",
    "src_id",
    "src_signal",
    "mic_id",
    "mic_pos_x",
    "mic_pos_y",
    "mic_pos_z",
    "room_rfl_west",
    "room_rfl_east",
    "room_rfl_north",
    "room_rfl_south",
    "room_rfl_ceiling",
    "room_rfl_floor",
]

subset_df = subset_df[needed_cols].reset_index(drop=True)
subset_df
```

## 4) Prepare HDF5 Inputs


```python
# 5) Build mic indices and design matrix for HDF5 extraction
# Clean numeric columns before saving/using


import re


clean_df = subset_df.copy()


def parse_fixed_8(v):
    s = str(v).strip()

    # if there are commas, normalize them (some exports use comma decimals)
    s = s.replace(",", ".")

    # keep only digits (remove all dots, spaces, etc.)
    digits = re.sub(r"[^0-9]", "", s)
    if digits == "":
        return np.nan

    # interpret as fixed-point with 8 decimal digits
    return int(digits) / 1e8


for c in ["mic_pos_x", "mic_pos_y", "mic_pos_z"]:
    clean_df[c] = clean_df[c].apply(parse_fixed_8).astype("float32")

clean_df[
    [
        "room_code",
        "src_id",
        "mic_id",
        "room_rfl_south",
        "room_rfl_north",
        "room_rfl_east",
        "room_rfl_west",
        "room_rfl_ceiling",
        "room_rfl_floor",
    ]
] = (
    clean_df[
        [
            "room_code",
            "src_id",
            "mic_id",
            "room_rfl_south",
            "room_rfl_north",
            "room_rfl_east",
            "room_rfl_west",
            "room_rfl_ceiling",
            "room_rfl_floor",
        ]
    ]
    .apply(pd.to_numeric)
    .astype("Int64")
)

clean_df = (
    clean_df.dropna().drop_duplicates().sort_values(by="mic_id").reset_index(drop=True)
)


print(f"Cleaned rows: {len(clean_df):,} (dropped {len(subset_df)-len(clean_df):,})")
```

```python
# 1. Define columns needed for the optimization objectives
POSITION_COLS = ['src_pos_x', 'src_pos_y', 'src_pos_z', 'mic_pos_x', 'mic_pos_y', 'mic_pos_z']
ROOM_ACOUSTIC_COLS = ['room_rfl_floor', 'room_rfl_ceiling', 'room_rfl_west', 'room_rfl_south', 'room_rfl_east', 'room_rfl_north']
SIGNAL_QUALITY_COLS = ['rec_silence_dB', 'rec_artifacts']

OBJECTIVE_COLS = [ 'room_code', 'src_id', 'mic_id', 'src_signal'] + POSITION_COLS + ROOM_ACOUSTIC_COLS + SIGNAL_QUALITY_COLS

# 2. Run Analysis Workflow
print("1. Validating positional data...")
metadata_df = validate_positional_data(metadata_df)

print("2. Computing Source-Mic distances...")
metadata_df = compute_src_mic_distance(metadata_df)

print("3. Filtering for specific 'chirp' signals...")
filtered_df = metadata_df[metadata_df['src_signal'] == 'chirp'].copy()

print(f"4. Selected {len(OBJECTIVE_COLS)} columns for {len(filtered_df)} samples.")
df_objectives = filtered_df[OBJECTIVE_COLS].copy()

```
